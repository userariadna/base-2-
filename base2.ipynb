```python
#no tomar encuenta si el estado es "RECHAZADA" o "NUEVO"
df.drop(df[df['Etapa'] == 'Rechazada' ].index, inplace = True)
df.drop(df[df['Etapa'] == 'Nuevo' ].index, inplace = True)
```

```python
df=df.rename(columns = {'Requirer Employee':'interventor'})
```

```python
def selector(x):   
    
    if 'HOSPEDAJE' in x:
        return 'hospedaje'
    if 'Hospedaje' in x:
        return 'hospedaje'
    if 'COMBUSTIBLE' in x:
        return 'gasolina'
    if 'GASOLINA' in x:
        return 'gasolina'
    if 'Interventoria' in x:
        return 'otros'
    if 'Gasolina' in x:
        return 'gasolina'
df['categorias'] = df['Título'].apply(selector)
```

```python
#suma por producto 
group = df.groupby(['categorias']).sum()
group.sort_values('Monto Requerido', ascending=True)
```

```python
#promedio de pago totales por interventor 
group_prom = df.groupby(['categorias']).mean()
group_prom.sort_values('Monto Requerido', ascending=True)
```

```python
#suma de pagos totales por interventor 
group_int= df.groupby(['interventor']).sum()
group_int.sort_values('Monto Requerido', ascending=True)
```
